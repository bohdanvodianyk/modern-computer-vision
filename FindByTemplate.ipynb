```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# imshow function
def imshow(title="Image", image=None, size=12):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
tamplate = cv2.imread("images/pharaoh.jpg")
imshow("Tamplate", tamplate)
```

```python
image = cv2.imread("images/find_waldo_full.jpg")
imshow("Where Pharaoh?", image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
tamplate = cv2.imread("images/pharaoh.jpg", 0)

result = cv2.matchTemplate(gray, tamplate, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

top_left = max_loc
bottom_right = (top_left[0]+40, top_left[1]+40)
cv2.rectangle(image, top_left, bottom_right, (0, 0, 255), 5)

imshow("Where Pharaoh?", image)
```
